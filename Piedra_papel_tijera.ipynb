```python
import random
import unicodedata


# --------------------------------
# Datos base
# --------------------------------

opciones = ("fuerza", "flexibilidad", "transformaciÃ³n")

micro_escenarios = [
    "En tu grupo hay una persona que no hace nada y otra que quiere hacerlo todo sola. El ambiente empieza a tensarse.",
    "Un compaÃ±ero hace un comentario sobre ti delante de la clase que te molesta, aunque dice que era una broma.",
    "En clase siempre hablan las mismas personas y te cuesta intervenir, aunque tienes ideas.",
    "Te enteras de que alguien ha contado algo tuyo que no querÃ­a que se supiera.",
    "El profesor cambia la fecha de un trabajo importante a Ãºltima hora y el grupo se queja.",
    "Dos personas discuten durante el recreo y te piden que tomes partido.",
    "Recibes un mensaje en un grupo de clase que te incomoda, pero nadie mÃ¡s parece darle importancia.",
    "Te equivocas al leer en voz alta y escuchas risas al fondo de la clase.",
    "En una actividad, sientes que el reparto de tareas no ha sido justo.",
    "TÃº piensas una cosa y la mayorÃ­a del grupo opina lo contrario, con bastante intensidad."
]

# Mensaje fijo de victoria para la ganadora
mensaje_ganadora = "âœ… Buena decisiÃ³n. Aunque todas las opciones son vÃ¡lidas, esta es posiblemente la mÃ¡s adecuada."

# Feedback para la no ganadora segÃºn elecciÃ³n
feedback_por_eleccion = {
    "fuerza": {
        "titulo": "ğŸ›¡ï¸ Poner lÃ­mites y defenderse",
        "texto": (
            "Has elegido una respuesta directa y firme.\n"
            "La fuerza emocional es Ãºtil para protegerte y marcar lÃ­mites.\n\n"
            "ğŸ§  Reflexiona:\n"
            "- Â¿Era necesario ser tan firme en esta situaciÃ³n?\n"
            "- Â¿CÃ³mo expresar la fuerza sin hacer daÃ±o?"
        )
    },
    "flexibilidad": {
        "titulo": "ğŸŒ¿ Adaptarse y escuchar",
        "texto": (
            "Has optado por una estrategia flexible.\n"
            "La flexibilidad ayuda a convivir y reducir conflictos.\n\n"
            "ğŸ§  Reflexiona:\n"
            "- Â¿He cedido porque querÃ­a o por evitar el problema?\n"
            "- Â¿QuÃ© lÃ­mites necesito cuidar?"
        )
    },
    "transformaciÃ³n": {
        "titulo": "ğŸŒ€ Cambiar la situaciÃ³n",
        "texto": (
            "Has elegido replantear la situaciÃ³n.\n"
            "La transformaciÃ³n busca nuevas formas de actuar.\n\n"
            "ğŸ§  Reflexiona:\n"
            "- Â¿QuÃ© idea nueva podrÃ­a cambiar el conflicto?\n"
            "- Â¿CuÃ¡ndo es buen momento para transformar?"
        )
    }
}

# --------------------------------
# Funciones auxiliares
# --------------------------------

def normalizar_input(texto):
    texto = texto.strip().casefold()
    texto = unicodedata.normalize("NFD", texto)
    return "".join(c for c in texto if unicodedata.category(c) != "Mn")

def mostrar_micro_escenario():
    escenario = random.choice(micro_escenarios)
    print("\nğŸ“˜ SITUACIÃ“N (lee y decide rÃ¡pido) âš¡ï¸")
    print(escenario)
    print()

def mostrar_feedback_ganadora(jugador):
    print(f"ğŸ’¬ Feedback para {jugador}")
    print(mensaje_ganadora)
    print()

def mostrar_feedback_eleccion(jugador, eleccion):
    feedback = feedback_por_eleccion[eleccion]
    print(f"ğŸ’¬ Feedback para {jugador}")
    print(feedback["titulo"])
    print(feedback["texto"])
    print()

def mostrar_feedback_final_grupal(historial):
    print("\nğŸ” FEEDBACK FINAL DEL GRUPO ğŸ“Œ")
    print(
        "Durante el juego habÃ©is utilizado distintas estrategias emocionales.\n"
        "Ninguna es mejor que otra: lo importante es aprender cuÃ¡ndo usar cada una.\n"
    )

    conteo = {"fuerza": 0, "flexibilidad": 0, "transformaciÃ³n": 0}
    for eleccion in historial:
        conteo[eleccion] += 1

    print("ğŸ“Š Estrategias utilizadas:")
    for estrategia, cantidad in conteo.items():
        print(f"- {estrategia.capitalize()}: {cantidad} veces")

    print(
        "\nğŸ“ Preguntas para el grupo:\n"
        "- Â¿QuÃ© estrategia usamos mÃ¡s?\n"
        "- Â¿CuÃ¡l nos cuesta mÃ¡s elegir?\n"
        "- Â¿En quÃ© situaciones reales podrÃ­amos probar otra diferente?"
    )

# --------------------------------
# FunciÃ³n para determinar ganadora
# --------------------------------

def determinar_ganadora(eleccion1, eleccion2):
    reglas = {
        "fuerza": "transformaciÃ³n",
        "flexibilidad": "fuerza",
        "transformaciÃ³n": "flexibilidad"
    }

    if eleccion1 == eleccion2:
        print(
            "ğŸ¤ Ambas jugadoras habÃ©is elegido la misma opciÃ³n. "
            "Â¡Bien! El consenso siempre es una buena estrategia."
        )
        return None

    if reglas[eleccion1] == eleccion2:
        return 1

    return 2

# --------------------------------
# Juego principal
# --------------------------------

def juego(max_rondas=3):
    historial = []

    print("Â¡Hola! Bienvenidas al juego ğŸ’ªFuerza, ğŸ¤¸â€â™€ï¸Flexibilidad y ğŸ’«TransformaciÃ³n\n")
    print("ğŸ® Este es un juego con el que poner a prueba diferentes estrategias de afrontaciÃ³n.\n")
    print(f"Se jugarÃ¡n {max_rondas} microescenarios y tendrÃ©is que seleccionar con que estrategia os sentÃ­s mÃ¡s identificadas.\n")
    print("Reglas express: elegÃ­s una estrategia. Si alguien escribe SALIR, se termina. ğŸ\n")

    jugador1 = input("ğŸ‘¤ Introduce tu nombre: ").strip().capitalize()
    jugador2 = input("ğŸ‘¤ Introduce tu nombre: ").strip().capitalize()

    for ronda in range(1, max_rondas + 1):
        print("\n" + "-" * 40)
        print(f"ğŸ”„ MICRO-ESCENARIO {ronda}")
        print("-" * 40)

        # Mostrar micro-escenario
        mostrar_micro_escenario()

        # Entrada vÃ¡lida para cada jugadora
        while True:
            eleccion1_raw = input(f"{jugador1}, elige FUERZA, FLEXIBILIDAD o TRANSFORMACIÃ“N (o SALIR): ")
            eleccion1_norm = normalizar_input(eleccion1_raw)
            if eleccion1_norm == "salir":
                print("ğŸ Juego finalizado por solicitud de salida.")
                return
            eleccion1 = "transformaciÃ³n" if eleccion1_norm == "transformacion" else eleccion1_norm
            if eleccion1 in opciones:
                break
            print("ğŸ¤” OpciÃ³n no vÃ¡lida. Â¡IntÃ©ntalo de nuevo!.")

        while True:
            eleccion2_raw = input(f"{jugador2}, elige FUERZA, FLEXIBILIDAD o TRANSFORMACIÃ“N (o SALIR): ")
            eleccion2_norm = normalizar_input(eleccion2_raw)
            if eleccion2_norm == "salir":
                print("ğŸ Juego finalizado por solicitud de salida.")
                return
            eleccion2 = "transformaciÃ³n" if eleccion2_norm == "transformacion" else eleccion2_norm
            if eleccion2 in opciones:
                break
            print("ğŸ¤” OpciÃ³n no vÃ¡lida. Â¡IntÃ©ntalo de nuevo!.")


        # Guardar elecciones
        historial.append(eleccion1)
        historial.append(eleccion2)

        # Determinar ganadora
        ganadora = determinar_ganadora(eleccion1, eleccion2)

        # Mostrar feedback
        if ganadora == 1:
            mostrar_feedback_ganadora(jugador1)
            mostrar_feedback_eleccion(jugador2, eleccion2)
        elif ganadora == 2:
            mostrar_feedback_ganadora(jugador2)
            mostrar_feedback_eleccion(jugador1, eleccion1)
        else:  # empate
            print("â¡ï¸ Siguiente microescenarioâ€¦\n")
            continue

    # Feedback final del grupo
    mostrar_feedback_final_grupal(historial)

# --------------------------------
# Ejecutar juego
# --------------------------------

# juego()
```
